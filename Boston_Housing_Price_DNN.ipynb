```
import tensorflow as tf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```
(train_features, train_labels), (test_features, test_labels) = tf.keras.datasets.boston_housing.load_data()
```

```
df = pd.DataFrame(train_features)
df.head(5)
```

```
train_mean = np.mean(train_features, axis=0)
train_std = np.std(train_features, axis=0)
train_normalized = (train_features - train_mean)/train_std
```

```
def create_model():
  model = tf.keras.Sequential(
      [tf.keras.layers.Dense(units=20, activation='relu', input_shape=[len(train_features[0])]),
       tf.keras.layers.Dense(units=20, activation='relu'),
       tf.keras.layers.Dense(units=1)]
  )
  model.compile(optimizer='adam', loss='mse', metrics=['mse', 'mae'])
  return model
```

```
model = create_model()
model.fit(train_normalized, train_labels, epochs=1000, validation_split=0.1)
```

```
test_features_normalized = (test_features - train_mean)/train_std
mse, _, _ = model.evaluate(test_features_normalized, test_labels)
rmse = np.sqrt(mse)
print('Root mean square error is : ', rmse)
```
